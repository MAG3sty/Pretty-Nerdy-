```python
#now testing chisquare for 'job_level' and 'salary'
```

```python
jb_crosstab = pd.crosstab(esi['job_level'], esi['salary'])
```

```python
jb_crosstab
```

```python
stats.chi2_contingency(jb_crosstab)
```

```python
#pvalue is 0.0 which is less than 0.5 meaning there is NOT a statistical significant; reject the null hypothesis
```

```python
#testing chisquare for 'dept' and 'salaries' 
```

```python
dept_crosstab = pd.crosstab(esi['Dept'], esi['salary'])
```

```python
dept_crosstab
```

```python
stats.chi2_contingency(dept_crosstab)
```
